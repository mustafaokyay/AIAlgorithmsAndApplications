#### c) Then, you try to predict whether a “competitor” is chocolate based on its other physical characteristics.
2. Build a classification algorithm.
3. Validate your results. How precise is your algorithm and what are its weaknesses?

```python
from sklearn.neighbors import KNeighborsClassifier 
import sklearn.model_selection as model_selection
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# I assumed the following subset (X) to be all of the physical features
X = candy[["fruity", "caramel", "peanutyalmondy", "nougat", "crispedricewafer", "hard", "bar", "pluribus"]]
y = candy ["chocolate"]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size = 0.4, random_state = 42)

knn=KNeighborsClassifier()
knn.fit(X_train, y_train)

knn_predict = knn.predict(X_test)

print(classification_report(y_test, knn_predict))
print(confusion_matrix(y_test, knn_predict))
print("\n")
print("Detailed Accuracy:")
print(accuracy_score(y_test, knn_predict))

# The trained model works quite well on the test set. In total, it only classified two actual chocolate candies
# as non-chocolate candies (as it can be seen from the confusion matrix)
```

#### d) Lastly, you want to understand the effect of the candy type on the win score.  You train a linear regression on all available features to predict the win percentage. Which weight does each characteristic have in your linear regression?

```python
from sklearn.linear_model import LinearRegression
import sklearn.model_selection as model_selection
from sklearn.metrics import mean_squared_error, explained_variance_score
from math import sqrt

X = candy.drop(['competitorname', 'winpercent'], axis=1) # create a dataframe without the label and any non-numeric column
y = candy ["winpercent"]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size = 0.4, random_state = 42)

linReg=LinearRegression()
linReg.fit(X_train, y_train)

linReg_predict = linReg.predict(X_test)

print("Root Mean Squared Error (RMSE):")
print(sqrt(mean_squared_error(y_test, linReg_predict)))
print("")
print("Computed weights:")
print(linReg.coef_)
print("for:")
print(X.columns)
```
